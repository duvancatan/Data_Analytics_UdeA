### Ambiente o Entorno Virtual en Python **venv** 

Un ambiente virtual en Python es un entorno aislado que permite instalar paquetes y dependencias sin afectar el resto del sistema ni otros proyectos. Es como crear una “Burbuja Independiente” de Python para cada proyecto.

Python, por diseño, instala las librerías de forma global. Esto genera un problema clásico: **distintos proyectos pueden necesitar versiones diferentes de la misma librería**, causando conflictos y errores. 
¡Los ambientes virtuales resuelven este problema!

#### Ejemplo 

Imagina dos proyectos:

- Proyecto A necesita: pandas 1.5

- Proyecto B necesita: pandas 2.2

- Sin ambientes virtuales → conflicto

- Con ambientes virtuales → cada proyecto funciona perfectamente.

#### Ventajas principales

1. *Evitan conflictos de dependencias*


- Cada proyecto tiene sus propias librerías.

- Esto garantiza que el código funcione siempre.



2. *Mejoran la reproducibilidad*


- Puedes compartir tu proyecto y otra persona puede ejecutarlo sin errores.

- Esto es crítico en Ciencia de Datos y Machine Learning.



3. *Organización profesional*

- Es el estándar en la industria.

- Empresas esperan que los proyectos estén aislados correctamente.



4. *Facilitan el despliegue*

- Herramientas como Docker funcionan basadas en este principio de aislamiento.

#### Casos de uso en Ciencia de Datos

1. *Entrenamiento de modelos*

Cada modelo puede requerir librerías específicas:

- scikit-learn

- TensorFlow

- PyTorch


2. *Desarrollo de aplicaciones de Machine Learning*

- APIs con FastAPI

- Apps con Streamlit

- Cada una con sus propias dependencias.


3. *Investigación y experimentación*

Permite probar nuevas librerías sin dañar proyectos existentes.


4. *Trabajo en equipos*

- Todos usan exactamente el mismo entorno.

- Esto evita el clásico problema:

- “En mi computador sí funciona”

## Tipos de Datos
Python tiene varios tipos de datos, los mismos datos básicos que otros lenguajes: enteros, flotantes, strings y booleanos. Además, las listas son un tipo predefinido en el lenguaje, los diccionarios, tuplas y sets.

La funcion que se utiliza para determinar el tipo de datos es *type()*

Para escribir comentarios es utiliza el caracter numeral *#*

### Tipos de datos integrados

En programación, el tipo de datos es un concepto importante.

Las variables pueden almacenar datos de diferentes tipos, y diferentes tipos pueden hacer diferentes cosas.

Python tiene los siguientes tipos de datos integrados de forma predeterminada, en estas categorías:

- Tipo de texto:	*str*
- Tipos numéricos:	*int, float, complex*
- Tipos de secuencia:	*list, tuple, range*
- Tipo de mapeo:	*dict*
- Tipos de conjuntos:	*set, frozenset*
- Tipo booleano:	*bool*
- Tipos binarios:	*bytes, bytearray, memoryview*
- Ninguno Tipo:	*NoneType*

```python
# str #
x = "Hello World"
type(x)
 
# int #
x = 20  
type(x)

# float #
x = 20.5
type(x)

# complex #
x = 1j  
type(x) 

# list #
x = ["apple", "banana", "cherry"]   
type(x)

# tuple #
x = ("apple", "banana", "cherry")   
type(x)

# range #
x = range(6)      
x
type(x)

# dict #
x = {"name" : "John", "age" : 36}      
type(x)

# set #
x = {"apple", "banana", "cherry"}       
type(x)

# bool #
x = True    
type(x)

# bytes #
x = b"Hello"    
type(x)

# bytearray #
x = bytearray(5)      
type(x)

# memoryview #
x = memoryview(bytes(5))     
type(x)

# NoneType #
x = None    
type(x)
```

### Nombre de variables
Al nombrar variables en Python, es importante seguir ciertas reglas y convenciones para garantizar que el código sea legible y funcione correctamente. Aquí hay algunas pautas clave para nombrar variables:
1. *Reglas de nombrado*
- Las variables deben comenzar con una letra (a-z, A-Z) o un guion bajo (_).
- No pueden comenzar con un número.
- No pueden contener espacios, pero se pueden usar guiones bajos para separar palabras (por ejemplo, `mi_variable`).
- No pueden ser palabras reservadas de Python (como `if`, `for`, `while`, etc.).    
2. *Convenciones de nombrado*
- Utiliza nombres descriptivos que indiquen claramente el propósito de la variable (por ejemplo, `edad_usuario` en lugar de `x`).
- Usa letras minúsculas y guiones bajos para separar palabras (snake_case) para mejorar la legibilidad (por ejemplo, `total_ventas`).
- Evita usar caracteres especiales o espacios en los nombres de variables.
- Sé consistente con el estilo de nombrado a lo largo de tu código para mejorar la legibilidad y el mantenimiento.
